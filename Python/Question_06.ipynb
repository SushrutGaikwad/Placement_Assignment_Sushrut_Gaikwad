# 4. All pokemons that do not have more than 2 `no_of_evolutions`:

```python
df[["prev_evolution", "next_evolution"]]
```

## Creating a new column called "`no_of_evolutions`":

```python
no_of_prev_evolutions = []
for i in range(len(df)):
    if type(df["prev_evolution"][i]) == str:
        lst = (df["prev_evolution"][i]).strip("][").split("}, {")
        no_of_prev_evolutions.append(len(lst))
    elif type(df["prev_evolution"][i]) == float:
        no_of_prev_evolutions.append(0) # considering NaN as 0 evolutions
```

```python
no_of_next_evolutions = []
for i in range(len(df)):
    if type(df["next_evolution"][i]) == str:
        lst = (df["next_evolution"][i]).strip("][").split("}, {")
        no_of_next_evolutions.append(len(lst))
    elif type(df["next_evolution"][i]) == float:
        no_of_next_evolutions.append(0) # considering NaN as 0 evolutions
```

```python
no_of_evolutions = np.add(no_of_prev_evolutions, no_of_next_evolutions) # total evolutions will be this sum
```

```python
df["no_of_evolutions"] = no_of_evolutions
df
```

## Filtering to get pokemons that do not have more than 2 `no_of_evolutions`:

```python
df[(df.no_of_evolutions <= 2)]
```

## Names of such pokemons:

```python
df[(df.no_of_evolutions <= 2)]["name"].unique()
```

# 5. All pokemons with `spawn_time` less than 300 seconds:

```python
df["spawn_time"]
```

```python
df["spawn_time"][0]
```

```python
type(df["spawn_time"][0])
```

## Creating a new column called "`spawn_time_sec`":

```python
spawn_time_sec = []

for i in range(len(df)):
    if type(df["spawn_time"][i]) == str:
        lst = df["spawn_time"][i].split(':')
        spawn_time_sec.append(int(lst[0])*60 + int(lst[1]))
    elif type(df["spawn_time"][i]) == float:
        spawn_time_sec.append(np.nan)
```

```python
df["spawn_time_sec"] = spawn_time_sec
df
```

## Filtering to get pokemons with less than 300 seconds of `spawn_time_sec`:

```python
df[(df.spawn_time_sec < 300)]
```

## Names of such pokemons:

```python
df[(df.spawn_time_sec < 300)]["name"].unique()
```
