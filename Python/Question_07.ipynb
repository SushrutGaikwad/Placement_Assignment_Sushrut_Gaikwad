```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Loading the dataset
df = pd.read_csv("Question_4_NASA_data.csv")
df.head()
```

```python
df.info()
```

# 1. All the Earth meteorites that fell before the year 2000:

## Creating a new column `only_year` that includes only the date in the format "YYYY":

```python
df["year"][0][:4]
```

```python
type(df["year"][0][:4])
```

```python
only_year = []
for i in range(len(df)):
    if type(df["year"][i]) == str:
        only_year.append(int(df["year"][i][:4]))
    elif type(df["year"][i]) == float:
        only_year.append(np.nan)
only_year
```

```python
df["only_year"] = only_year
df.head()
```

## Filtering to get all the Earth meteorites that fell before the year 2000:

```python
df[(df.only_year < 2000)]
```

## Name of such meteorites:

```python
df[(df.only_year < 2000)]["name"].unique()
```

# 2. Coordinates of all the Earth meteorites that fell before the year 1970:

```python
df[(df.only_year < 1970)]["geolocation"]
```

## Creating a new column `coordinates` utilizing the already existing column `geolocation`:

```python
coordinates = []
for i in range(len(df)):
    if type(df["geolocation"][i]) == str:
        coordinates.append(df["geolocation"][i].split(":")[-1].strip("}").strip(" "))
    elif type(df["geolocation"][i]) == float:
        coordinates.append(np.nan)
```

```python
df["coordinates"] = coordinates
df.head()
```

## Filtering to get the `coordinates`:

```python
df[(df.only_year < 1970)]["coordinates"]
```

# 3. All meteorites with `mass` greater than 10000 kg:

```python
df[(df.mass > 10000)]
```

## Names of such meteorites:

```python
df[(df.mass > 10000)]["name"].unique()
```
