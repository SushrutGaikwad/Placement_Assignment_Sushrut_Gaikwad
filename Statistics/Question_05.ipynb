**Question 5: A certain city has two taxi companies: Company A has 80% of the taxis and Company B has 20% of the taxis. Company A's taxis have a 95% success rate for picking up passengers on time, while Company B's taxis have a 90% success rate. If a randomly selected taxi is late, what is the probability that it belongs to Company A?**

**Answer:**

Let $A$ denote the event that the taxi belongs to company A, $B$ denote the event that the taxi belongs to company B, $S$ denote the event of successfully picking up passengers on time, and $F$ denote the event of failing to pick up passengers on time. We are asked to find $P\left( A \:\middle|\: F \right)$. Using the formula for conditional probability, this can be written as $$P\left( A \:\middle|\: F \right) = \dfrac{P \left( A \cap F \right)}{P \left( F \right)}$$
We can find the numerator and the denominator from the above tree diagram, giving us $$P\left( A \:\middle|\: F \right) = \dfrac{0.8\times 0.05}{\left( 0.8\times 0.05 \right) + \left( 0.2\times0.1 \right)}\approx0.67$$
