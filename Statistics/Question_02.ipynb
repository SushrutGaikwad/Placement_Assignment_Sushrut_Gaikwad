**Question 2: Consider a dataset containing the heights (in centimeters) of 1000 individuals. The mean height is 170 cm with a standard deviation of 10 cm. The dataset is approximately normally distributed, and its skewness is approximately zero. Based on this information, answer the following questions:**
1. What percentage of individuals in the dataset have heights between 160 cm and 180 cm?
2. If we randomly select 100 individuals from the dataset, what is the probability that their average height is greater than 175 cm?
3. Assuming the dataset follows a normal distribution, what is the z-score corresponding to a height of 185 cm?
4. We know that 5% of the dataset has heights below a certain value. What is the approximate height corresponding to this threshold?
5. Calculate the coefficient of variation (CV) for the dataset.
6. Calculate the skewness of the dataset and interpret the result.

**Answers:**

The following is what a normal distribution looks like.

For the given data, $\mu = 170$ cm, and $\sigma = 10$ cm.

**Answer 1:**

We have $160 = \mu - \sigma$ and $180 = \mu + \sigma$. From the plot, it is clear that there are approximately $34.1\% + 34.1\% = 68.2\%$ people between this range.

**Answer 2:**

According to the central limit theorem, randomly sampling a variable following a normal distribution with a mean of $\mu$ and standard deviation $\sigma$ will also give normally distributed values with a sample mean $\mu_s = \mu$ and a standard deviation of $\sigma_s = \dfrac{\sigma}{\sqrt{n}}$. For this example, we have $n=100$, which means $\mu_s=170$ and $\sigma_s=\dfrac{10}{\sqrt{100}}=1$. Further, notice that $175=\mu_s+5\sigma_s$. According to the above plot, the probability that the height is above $175$ is less than $1\%$.

**Answer 3:**

The z-score is given by $$z=\dfrac{x-\mu}{\sigma}$$

where, $x$ is the value of the variable. So, the z-score for $x=185$ is $$z=\dfrac{185-170}{10}=1.5$$

**Answer 4:**

For this, we will use the negative z-score table. The z-score value corresponding to $0.05$ is $-1.64$. So, we have $$-1.64 = \dfrac{x-170}{10}$$

Solving for $x$, we get $$x=153.6$$

**Answer 5:**

The coefficient of variation $\left( \text{CV} \right)$ is given by $$\text{CV} = \dfrac{\sigma}{\mu} = \dfrac{10}{170} \approx 0.059$$

**Answer 5:**

Skewness is given by $$\text{Skweness}=\dfrac{\sum_{i=1}^{N} \left( x_i - \mu \right)^3}{\left( N-1 \right)\times \sigma^3}$$

We know the values $N=1000$, $\mu=170$, $\sigma=10$, but we do not know the individual values of the random variable, i.e., $x_i$. Hence, we cannot calculate the skewness. Further, it is mentioned in the question that the dataset is normally distributed, and its skewness is approximately zero.
